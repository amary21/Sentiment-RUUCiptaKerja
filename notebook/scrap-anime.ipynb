```python
import re
import csv
from getpass import getpass
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
# from msedge.selenium_tools import Edge, EdgeOptions
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

# navigate to login screen
driver.get('https://otakudesu.moe/wpoiec-episode-1-sub-indo/')
# driver.maximize_window()
```

```python
# responsive-embed-stream
```

```python
# Episode Selanjutnya
```

```python
data = []
for x in range(983):
    res_video = driver.find_element_by_xpath('//div[@class="responsive-embed-stream"]/iframe')
    video = res_video.get_attribute('src')
    title = driver.find_element_by_class_name('posttl').text
    eps = (title, video)
    data.append(eps)
    print(eps)
    driver.find_element_by_link_text('Next Eps.').click()

driver.close()
```

```python
with open('onepiece.csv', 'w', newline='', encoding='utf-8') as f:
    header = ['eps_title', 'video']
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data)
```
